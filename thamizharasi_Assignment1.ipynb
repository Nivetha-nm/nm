# Visualization
# univariate analysis

```python
import seaborn as sns
```

box plot

```python
sns.boxplot(y=df['Price'])
```

violin plot

```python
sns.violinplot(y=df['Price'])
```

strip plot

```python
sns.stripplot(y=df['Price'])
```

count plot

```python
df['grade of the house']
```

```python
sns.countplot(x=df['grade of the house'])
```

# bivariate analysis
line plot

```python
sns.lineplot(data=df,x='number of floors',y='Price')
```

bar plot

```python
sns.barplot(data=df,x='grade of the house',y='Price')
```

scatter plot

```python
sns.scatterplot(data=df,x='Area of the house(excluding basement)',y='Price')
```

boxen plot

```python
sns.boxenplot(data=df,x="lot area",y="Price")
```

dist plot

```python
sns.distplot(df['Price'])
```

joint plot

```python
sns.jointplot(data=df,x='grade of the house',y='Price')
```

kde plot

```python
sns.kdeplot(data=df,x='grade of the house',y='Price')
```
